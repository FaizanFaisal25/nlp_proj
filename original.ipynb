```python
import glob
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
import sys
```

```python
loss = []
for i in glob.glob("./Original, lr e-3/test/*.txt"):
    if "10000" in i:
        continue
    print(i)
    with open(i) as f:
        line = f.readlines()[1]
        line = line.split("=")
        loss.append(float(line[1]))

with open("./Original, lr e-3/test/test-10000.txt") as f:
    line = f.readlines()[1]
    line = line.split("=")
    loss.append(float(line[1]))

print(loss)
plt.plot(loss)
```
